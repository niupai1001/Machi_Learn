```python
import numpy as np

"""
ç®—æ³•æ­¥éª¤
è¾“å…¥:çº¿æ€§å¯åˆ†çš„æ•°æ®é›†T,i=1,2,â€¦,N;å­¦ä¹ ç‡n(0<nâ‰¤1); è¾“å‡º: a,b
(1) a = 0,b = 0; 
(2)è®¡ç®—xiå’Œxjçš„å†…ç§¯çŸ©é˜µGram_matrix(å¯¹å¶å½¢å¼èŠ‚çœæ—¶é—´çš„å…³é”®æ‰€åœ¨)
(3)åœ¨è®­ç»ƒé›†ä¸­é€‰å–æ•°æ®(xi,yi); 
(4)å¦‚æœyi([âˆ‘j->N aj*yj*xjÂ·xi] + b) <= 0 , ai += n(ç¬¬iä¸ªå®ä¾‹ç‚¹çš„è¿­ä»£æ¬¡æ•°å¢åŠ ä¸€ä¸ªå­¦ä¹ ç‡)  b += n*yi
(5)è½¬è‡³(3)ç›´åˆ°æ²¡æœ‰è¯¯åˆ†ç±»æ•°æ®ã€‚

"""


def update(item, i):    # æ›´æ–°å®ä¾‹ç‚¹
    global alpha, b
    alpha[i][0] += l    # ç¬¬iä¸ªå®ä¾‹ç‚¹çš„è¿­ä»£æ¬¡æ•°å¢åŠ ä¸€ä¸ªå­¦ä¹ ç‡
    b += l*item[-1]
    print(alpha, b)


def check(_GramMatrix):  # å¯¹åº”ç®—æ³•ç¬¬å››æ­¥
    separated_flag = True  # é»˜è®¤æ— é”™è¯¯åˆ†ç±»ç‚¹
    res = 0
    i = 0
    for train_item in train_set:
        yi = train_item[-1]
        for j in range(train_set.shape[0]):
            yj = train_set[j][-1]   # ç¬¬jä¸ªå®ä¾‹ç‚¹
            aj = alpha[j][0]    # ç¬¬jä¸ªå®ä¾‹ç‚¹çš„è¿­ä»£æ¬¡æ•°
            # è®¡ç®—âˆ‘ï¼ˆj->Nï¼‰ aj* yj * xjÂ·xi
            res += aj * yj * _GramMatrix[i][j]
        # print(res)
        # è®¡ç®— (ğ‘¦(ğ‘–)*âˆ‘alpha(j)*ğ‘¦(ğ‘—)ï¼‰*ğ‘¥(ğ‘—)âˆ™ğ‘¥(ğ‘–)â‰¤0
        if (res + b) * yi <= 0:
            separated_flag = False
            update(train_item, i)
        res = 0
        i += 1
    return separated_flag


def get_GramMatrix(_train_set):  # è®¡ç®—å†…ç§¯çŸ©é˜µ
    shape = _train_set.shape
    matrix = np.zeros((shape[0] ,shape[0]))

    item_list = []
    for train_item in _train_set:
        item_list.append(train_item[:-1])

    for i in range(shape[0]):
        for j in range(shape[0]):
            matrix[i][j] = np.vdot(item_list[i], item_list[j])

    print(matrix)
    return matrix


if __name__ == '__main__':
    train_set = np.array([[3, 3, 1],  # è®¾ç½®è®­ç»ƒé›†
                          [4, 3, 1],
                          [1, 1, -1]])
    # train_set1 = np.random.normal(3, size=(10, 3))
    # train_set2 = np.random.normal(7, size=(10, 3))
    # train_set = np.concatenate((train_set1, train_set2), axis=0)
    # è®¾ç½®åˆå€¼
    l = 1  # å­¦ä¹ ç‡
    alpha = np.zeros((train_set.shape[0], 1), int) * l  # å½“å­¦ä¹ ç‡ï¼ˆæ­¥é•¿ï¼‰ä¸º1æ—¶ï¼Œalphaå°±æ˜¯æ­Œå®ä¾‹ç‚¹æ›´æ–°çš„æ¬¡æ•°
    b = 0  # åç½®
    w = 0  # æƒå€¼

    GramMatrix = get_GramMatrix(train_set)
    for i in range(50):  # è¿­ä»£50æ¬¡
        if check(GramMatrix):
            print("ç¬¬{}æ¬¡å®Œæˆå…¨éƒ¨åˆ†ç¦»".format(i))
            break

    for i, item in enumerate(train_set):
        w += alpha[i][0] * item[-1] * item[:-1]

    print("å‚æ•°ä¸ºï¼šw = {}, b = {}".format(w, b))


```
